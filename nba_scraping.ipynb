```python
from datetime import datetime, timedelta

start_date_1 = datetime(2000, 1, 1)
end_date_1 = datetime(2009, 12, 31)
start_date = datetime(2010, 1, 1)
end_date  = datetime(2023, 12, 31)

seasons = []
while start_date_1 <= end_date_1:
    year = start_date_1.strftime("%Y")
    short_year = str(int(start_date_1.strftime("%y")) + 1)
    year_format = year + "-" +"0"+short_year
    seasons.append(year_format)
    start_date_1 += timedelta(days=365)

while start_date <= end_date:
    year =start_date.strftime("%Y")
    short_year = str(int(start_date.strftime("%y")) + 1)
    year_format = year + "-" +short_year
    seasons.append(year_format)
    start_date += timedelta(days=365)

print(seasons)
```

```python
import os

def check_folder_in_jupyter(folder_name):
    current_path = os.getcwd()
    for root, dirs, files in os.walk(current_path):
        if folder_name in dirs:
            return True
    return False



def check_file_in_folder(folder_path, file_name):
    for root, dirs, files in os.walk(folder_path):
        if file_name in files:
            return True
    return False
```

```python
#pull the Table for other seasons

for season in seasons:
    if check_file_in_folder("nba_stars", "{}.html".format(season)) is False:
        url = stars_url.format(season)
        driver = webdriver.Chrome()
        driver.get(url)
        time.sleep(3)
        html = driver.page_source
        
        with open("nba_stars/{}.html".format(season), "w+", encoding= 'utf-8') as f:
            f.write(html)
```

```python
dfs =[]
for season in seasons:
    with open("nba_stars/{}.html".format(season)) as f:
        page = f.read()
    
    soup = BeautifulSoup(page, 'html.parser')
    stars_tables = soup.select ('table', class_ = "Crom_base__f0niE")
    stars = pd.read_html(StringIO(str(stars_tables)), encoding= 'utf-8')[-1]
    
    s = str(season).split("-")[1]
    s_1 = str(season).split("-")[0][:2]
    stars["year"] = s_1 + s

    stars.columns = [c.lower() for c in stars.columns]
    stars.columns

    #stars = stars[["player", "team", "year", "age","gp","min", "w", "l", "pts", "3pm", "ast","fgm", "fga", "ftm" ]]
    dfs.append(stars)
```

```python
df = pd.concat(dfs)
df = df[["player", "team", "year", "age","gp","min", "w", "l", "pts", "3pm", "ast","fgm", "fga", "ftm" ]]
```

```python
df.to_csv("nba_players.csv", index= False)
```
