```python
df_com = {"computer": computer_data, "speaker": speakers_data,
         "keyboard": keyboard_data, "laptop": laptops_data, 
         "power_bank": power_bank_data, "television": television_data,
         "men_watches": men_watches_data}
```

```python
for name, df in df_com.items():
    print(f"Working on {name.upper()} data.......")
    
    def string_list(image_url):
        for i,link in enumerate(image_url):
            if len(link) > 0:
                image_url[i] = link[2:-2] #used str(link[2:-2]).split(" ")[0] for duplicate image_url
            else:
                image_url[i]  = None
        return image_url

    
    df['image_url'] = string_list(df['image_url'])
    df['product_details'] = string_list(df['product_details'])
    df['available_option'] = string_list(df['available_option'])

    df['avg_rating'] = df['avg_rating'].str.replace(r"[()]", "", regex=True)
    df["number_of_rating"] = df['number_of_rating'].apply(lambda x: str(x).split()[0])
    df[['full_price', "discount_price"]] = df[['full_price', "discount_price"]].replace(r"\$", "", regex=True)

    convert_columns = ["full_price", "discount_price", "avg_rating", "number_of_rating"]
    df[convert_columns] =  df[convert_columns].apply(pd.to_numeric, errors = 'coerce')
    df = df.fillna(0)

    duplicates  = df[df.duplicated()]
    print(f"Found {len(duplicates)} duplicate OUT OF {len(df)} {name} data")
    df = df.drop_duplicates(keep="last")

    print(f"Now Saving cleaned {name.upper()} data to File")
    df.to_csv(f"walmart_gadgets/{name}_products.csv", index=False,  encoding="utf-8")

    print("Unto The Next DataFrame..............")
```
