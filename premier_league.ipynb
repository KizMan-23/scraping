```python
years = list(range(2022,2019, -1))
```

```python
years
```

```python
standing_url = "https://fbref.com/en/comps/9/Premier-League-Stats"
```

```python
all_matches = []

for year in years:
    driver = webdriver.Chrome()
    driver.get(standing_url)
    time.sleep(2)
    html = driver.page_source
    
    with open("premier_league/{}.html".format(year), "w+", encoding='utf-8') as f:
        f.write(html)

    soup = BeautifulSoup(html, "html.parser")
    standn = soup.select("table.stats_table")
    links = [l.get('href') for l in standn[0].find_all('a', href = True)]
    links = [l for l in links if '/squads/' in l]
    team_urls  = [f"https://fbref.com{l}" for l in links]
    print("Links Successful")

    previous_season = soup.select("a.prev")[0].get("href")
    standing_url = f"https://fbref.com{previous_season}"
    
    for team_url in team_urls:
        team_name = team_url.split("/")[-1].replace("-Stats", "").replace("-"," ")
        data = requests.get(team_url)
        matches = pd.read_html(StringIO(str(data.text)), match="Scores & Fixtures")[0]

        parsed_page = BeautifulSoup(data.text, "html.parser")
        links = [a['href'] for a in parsed_page.find_all('a', href=True)]
        links = [l for l in links if l and '/all_comps/shooting/' in l]
        
        data = requests.get(f"https://fbref.com{links[0]}")
        shooting = pd.read_html(StringIO(data.text), match= "Shooting")[0]
        shooting.columns = shooting.columns.droplevel()

        try:
            team_data = matches.merge(shooting[["Date", "Time", "SoT", "Dist", "PK", "FK", "PKatt"]], on =("Date" and "Time"))
        except ValueError:
            continue

        team_data = team_data[team_data["Comp"] == "Premier League"]
        team_data['Team'] = team_name
        team_data['Year'] = year
        del team_data['Date_y']
        
        all_matches.append(team_data)
```

```python
match_df = pd.concat(all_matches)
```

```python
match_df.columns = [c.lower() for c in match_df.columns]
```

```python
match_df
```

```python
match_df.to_csv("premier_league_matches.csv", index=False)
```
