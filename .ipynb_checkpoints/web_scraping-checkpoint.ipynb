```python
years = list(range(1992,2023))
years
```

```python
url_str = "https://www.basketball-reference.com/awards/awards_{}.html"
```

```python
import requests
for year in years:
    url = url_str.format(year)
    data = requests.get(url)

    with open("mvps/{}.html".format(year), "w+", encoding='utf-8') as f:
        f.write(data.text)
```

```python
from bs4 import BeautifulSoup
```

```python
with open("mvps/1992.html",encoding='utf-8') as f:
    page = f.read()
```

```python
soup = BeautifulSoup(page, "lxml")
soup.find("tr", class_="over_header").decompose()
```

```python
mvp_table = soup.select("table", class_ = "sortable stats_table")
```

```python
mvp_table[0]
```

```python
import pandas as pd 
#the string function of the read_html is deprecated and needs StringIO
from io import StringIO

mvp_fr = pd.read_html(StringIO(str(mvp_table)),encoding='utf-8')[0]
mvp_fr["Year"] = years[1]
mvp_fr
```

```python
dfs = []
for year in years:
    with open("mvps/{}.html".format(year),encoding='utf-8') as f:
        page = f.read()
    soup = BeautifulSoup(page, "lxml")
    soup.find("tr", class_="over_header").decompose()
    mvp_table = soup.find_all(id="mvp")

    mvp_stand = pd.read_html(StringIO(str(mvp_table)),encoding='utf-8')[0]
    mvp_stand["Year"] = year

    dfs.append(mvp_stand)
```

```python
df = pd.concat(dfs)
df
```

```python
df.to_csv("mvps_table.csv")
```

Downloading Players Stat

```python
player_url = "https://www.basketball-reference.com/leagues/NBA_{}_per_game.html"
```

```python
#check for files contained in the folders already
import os

def check_folder_in_jupyter_lab(folder_name):
    current_path = os.getcwd()
    for root, dirs, files in os.walk(current_path):
        if folder_name in dirs:
            return True
    return False



def check_file_in_folder(folder_path, file_name):
    for root, dirs, files in os.walk(folder_path):
        if file_name in files:
            return True
    return False

```

```python
#The page contains a long table run with JS.
#use selenium to render the full JS table
from selenium import webdriver
import time


for year in years:
    if check_file_in_folder("players", "{}.html".format(year)) is False:
        url = player_url.format(year)
        driver = webdriver.Chrome()
        
        driver.get(url)
        time.sleep(2)
        html = driver.page_source
        
        with open("players/{}.html".format(year), "w+", encoding= 'utf-8') as f:
            f.write(html)
```

```python
dfs = []
for year in years:
    with open("players/{}.html".format(year),encoding='utf-8') as f:
        page = f.read()
    soup = BeautifulSoup(page, "lxml")
    players_tab =soup.find_all(id = "per_game_stats")
    players_table = pd.read_html(StringIO(str(players_tab)), encoding= 'utf-8')[0]
    players_table["Year"] = year

    dfs.append(players_table)
```

```python
player_df = pd.concat(dfs)
```

```python
player_df.to_csv("players.csv", index= False)
```

Downloading Team Stats

```python
team_url = "https://www.basketball-reference.com/leagues/NBA_{}_standings.html"
```

```python
import requests

for year in years:
    url = team_url.format(year)
    data = requests.get(url)
    
    with open("teams/{}.html".format(year), 'w+', encoding='utf-8') as f:
        f.write(data.text)
```

```python
dfs = []

for year in years:
    with open("teams/{}.html".format(year), encoding='utf-8') as f:
        page = f.read()
    soup = BeautifulSoup(page, 'html.parser')
    soup.find("tr", class_ = "thead").decompose()
    team_tab = soup.find_all(id="divs_standings_E")
    team = pd.read_html(StringIO(str(team_tab)),encoding='utf-8')[0]
    team["Year"] = year
    team["Team"] = team["Eastern Conference"]
    del team["Eastern Conference"]
    dfs.append(team)
    
    
    soup = BeautifulSoup(page, 'html.parser')
    soup.find("tr", class_ = "thead").decompose()
    team_tab = soup.find_all(id="divs_standings_W")
    team = pd.read_html(StringIO(str(team_tab)),encoding='utf-8')[0]
    team["Year"] = year
    team["Team"] = team["Western Conference"]
    del team["Western Conference"]
    dfs.append(team)
```

```python
df = pd.concat(dfs)
```
